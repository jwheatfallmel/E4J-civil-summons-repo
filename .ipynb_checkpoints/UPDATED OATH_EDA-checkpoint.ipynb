```python
# Binning time by early morning (00-05); Morning (06-11); Afternoon (12-16); Evening (12-20); Night (21-23) 

# Function for binning hours 
def time_of_day(hour):
    if pd.isna(hour):
        return None
    if 0 <= hour < 6:
        return "Early Morning"
    elif 6 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

civil_summons_2021['O_Violation_Time_Category'] = civil_summons_2021['Violation_Hour'].apply(time_of_day)
```

#### "O_Payment_Status" <- O_Penalty_Imposed: The actual penalty the respondent is required to pay & O_Paid_Amount: The total amount paid by the respondent

ANALYSIS: Different from hearing status, directly indicates willingness to pay. Created a new variable "Payment-Status" which shows that 2519 had no penalty imposed, 1152 unpaid, 853 paid in full; 62 missing data; 58 overpaid/adjustment; 19 partially paid

```python
# Making sure all are numeric 
civil_summons_2021['O_Penalty_Imposed'] = pd.to_numeric(
    civil_summons_2021['O_Penalty_Imposed'], errors='coerce'
)

civil_summons_2021['O_Paid_Amount'] = pd.to_numeric(
    civil_summons_2021['O_Paid_Amount'], errors='coerce'
)

civil_summons_2021['Amount_Difference'] = (
    civil_summons_2021['O_Penalty_Imposed'] - civil_summons_2021['O_Paid_Amount']
)
```

```python
# Function for binning into paid categories
def payment_status(row):
    if pd.isna(row['O_Penalty_Imposed']) or pd.isna(row['O_Paid_Amount']):
        return "Missing Data"
    if row['O_Penalty_Imposed'] == 0:
        return "No Penalty Imposed"
    if row['O_Paid_Amount'] == 0:
        return "Unpaid"
    if row['O_Paid_Amount'] == row['O_Penalty_Imposed']:
        return "Paid in Full"
    if row['O_Paid_Amount'] < row['O_Penalty_Imposed']:
        return "Partially Paid"
    if row['O_Paid_Amount'] > row['O_Penalty_Imposed']:
        return "Overpaid / Adjustment"
        
civil_summons_2021['O_Payment_Status'] = civil_summons_2021.apply(payment_status, axis=1)
```

```python
# Making final_discipline_charge
charge_cols = ['DA_charge', 'CC_charge']

def combine_all_charges(row):
    charges = []

    for col in charge_cols:
        value = row[col]
        if value not in [".", None, "", " "] and pd.notna(value):
            charges.append(value.strip())

    # remove duplicates, preserve order
    charges = list(dict.fromkeys(charges))

    # return "." if no real charges
    if len(charges) == 0:
        return "."

    return "; ".join(charges)

civil_summons_2021['Full_Charge'] = civil_summons_2021.apply(combine_all_charges, axis=1)
```

```python
def bin_charge_3(charge):
    charge = str(charge).upper()

    if (
        'RIGHT OF WAY' in charge or
        'FAILURE TO YIELD' in charge or
        'VEHICLE' in charge or
        'BICYCLE' in charge or
        'SCOOTER' in charge or
        'MOTOR' in charge or
        'PEDESTRIAN' in charge or
        'PHYSICAL INJURY' in charge
    ):
        return 'Charge_Public_Safety'

    elif (
        'VEND' in charge or
        'MFVU' in charge or
        'URINATION' in charge or
        'LITTER' in charge or
        'SPITTING' in charge or
        'LICENSE' in charge or
        'PLATE' in charge or
        'PERMIT' in charge or
        'SIDEWALK' in charge or
        'OBSTRUCT' in charge or
        'DISPLAY' in charge or
        'NOISE' in charge or
        'PARK' in charge or
        'SMOK' in charge or 
        'DUMP' in charge
    ):
        return 'Charge_Quality_of_Life'

    else:
        return 'Charge_Administrative'
```

```python
# Necessary loadings 
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
aa_columns = [
    'roster_rank',
    'roster_race',
    'roster_gender',
    'has_award',
    'training_count_bin',
    'Charge_Category',
    'Violation_Severity_Tier',
    'O_Violation_Location_Borough',
    'O_Violation_Time_Category',
    'Phenotype',
    'O_Payment_Status',
    'O_Compliance_Status',
    'O_Hearing_Status'
]


aa_df = civil_summons_2021[aa_columns].copy()
```

```python
# 2,020 frequent itemsets will be generated with a support threshold of 10%. 
# Meaning that 2,020 combination of officer behavior/enforcement attributes occur in at least 10% of the 4,663 tickets 
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
len(frequent_itemsets)
```

```python
# USING FOR ANALYSIS 
# Made lift 1.1 for meaningful itemsets 
rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1.1
)
```
